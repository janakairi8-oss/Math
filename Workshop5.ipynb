```python
import numpy as np
import matplotlib.pyplot as plt
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
# Add vectors V and W to the plot
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
U = np.array([1,0.25,0])
V = np.array([0,1,0.25])
W = np.array([0.25,0,1])
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
origin = [0,0,0]
ax.quiver(*origin, U[0],U[1],U[2])
ax.quiver(*origin, V[0],V[1],V[2])
```

```python
import numpy as np
def magnitude(vector):
  return np.sqrt(np.sum(vector**2))
U = np.array([1.5,1.5,2.0])
mag = magnitude(U)
print("magnitude of U is :",mag)
```

```python
import numpy as np
def direction_cosines(vector):
  magnitude = np.sqrt(np.sum(vector**2))
  return vector / magnitude
O = np.array([1.5,1.5,2.0])
dc = direction_cosines(O)
print("direction_cosines is:",dc)
```

```python
import numpy as np
def dot_product(a,b):
  return np.sum(a*b)
a = np.array([1.5,1.5,2.0])
b = np.array([-1.5,0.5,2.0])
result = dot_product(a,b)
print("dot product:",result)
```

```python
import numpy as np
def cross_product(a,b):
  return np.cross(a,b)
a = np.array([1.5,1.5,2.0])
b = np.array([-1.5,0.5,2.0])
result = cross_product(a,b)
print("cross product:",result)
```
